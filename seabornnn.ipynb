```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

tips
```

```python
# Scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Scatter Plot with Hue (Lunch/Dinner)")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", size="size", style="time", sizes=(20, 200))
plt.title("Scatter Plot with Size, Style, and Hue")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", palette="rocket")
plt.title("Scatter Plot with Custom Palette")
plt.show()
```

```python
#lineplot

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
fmri = sns.load_dataset("fmri")

fmri
```

```python
#bar plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Basic bar chart
sns.barplot(data=tips, x="day", y="total_bill")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Average Total Bill by Day and Gender")
plt.show()
```

```python
from numpy import median

sns.barplot(data=tips, x="day", y="total_bill", estimator=median, ci=None)
plt.title("Median Total Bill by Day")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", palette="pastel", ci=None)
plt.title("Bar Chart with Custom Palette")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", order=["Thur", "Fri", "Sat", "Sun"])
plt.title("Bar Chart with Ordered Days")
plt.show()
```

```python
sns.barplot(data=tips, x="total_bill", y="day", ci=None)
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
#histogram
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Basic histogram
sns.histplot(data=tips, x="total_bill", bins=10)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", hue="sex", bins=20, palette="pastel")
plt.title("Distribution of Total Bill by Gender")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Histogram with KDE")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", hue="sex", multiple="stack", bins=20)
plt.title("Stacked Histogram by Gender")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", y="tip", bins=20, cbar=True)
plt.title("2D Histogram of Total Bill and Tip")
plt.show()
```

```python
#pairplot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
iris = sns.load_dataset("iris")

# Basic pair plot
sns.pairplot(data=iris)
plt.show()
```

```python
sns.pairplot(data=iris, hue="species", palette="bright")
plt.show()
```

```python
sns.pairplot(data=iris, hue="species", diag_kind="kde")
plt.show()
```

```python
sns.pairplot(data=iris, hue="species", kind="kde")
plt.show()
```

```python
sns.pairplot(data=iris, vars=["sepal_length", "sepal_width", "petal_length"], hue="species")
plt.show()
```

```python
sns.pairplot(data=iris, hue="species", corner=True)
plt.show()
```

```python
sns.pairplot(data=iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
#boxplot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Basic box plot
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Box Plot with Hue (Gender)")
plt.show()
```

```python
sns.boxplot(data=tips, x="total_bill", y="day", orient="h")
plt.title("Horizontal Box Plot")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", palette="Set2")
plt.title("Box Plot with Custom Palette")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", notch=True, palette="coolwarm")
plt.title("Box Plot with Notches")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", showfliers=False)
plt.title("Box Plot without Outliers")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="time", dodge=True)
plt.title("Grouped Box Plot by Time")
plt.show()
```

```python
#heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.rand(6, 6)

# Basic heatmap
sns.heatmap(data)
plt.title("Basic Heatmap")
plt.show()
```

```python
sns.heatmap(data, vmin=0.2, vmax=0.8, cmap="viridis")
plt.title("Heatmap with Custom Color Range")
plt.show()
```

```python
sns.heatmap(data, linewidths=0.5, linecolor="white", cmap="magma")
plt.title("Heatmap with Line Separators")
plt.show()
```

```python
sns.heatmap(data, square=True, cmap="Blues")
plt.title("Heatmap with Square Cells")
plt.show()
```

```python
sns.heatmap(data, cbar=False, cmap="cubehelix")
plt.title("Heatmap without Color Bar")
plt.show()
```

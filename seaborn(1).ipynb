```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
df = sns.load_dataset('tips')

# Display the first few rows of the dataframe
df
```

```python
# Scatter plot showing the relationship between total bill and tip amount, with color based on the day
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')

# Display the plot
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
# Bar plot showing the average total bill by day of the week
sns.barplot(x='day', y='total_bill', data=df)

# Display the plot
plt.title("Average Total Bill by Day")
plt.show()
```

```python
# Box plot showing the distribution of total bill across different days
sns.boxplot(x='day', y='total_bill', data=df)

# Display the plot
plt.title("Box Plot: Total Bill Distribution by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
df = sns.load_dataset('tips')

# Create a KDE plot for the 'total_bill' variable
sns.kdeplot(df['total_bill'], shade=True, color='blue')

# Add title and labels
plt.title('Kernel Density Estimate (KDE) of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')

# Display the plot
plt.show()
```

```python
# Violin plot showing the distribution of total bill for each day
sns.violinplot(x='day', y='total_bill', data=df)

# Display the plot
plt.title("Violin Plot: Total Bill Distribution by Day")
plt.show()
```

```python
# Pairplot to show pairwise relationships, colored by 'sex'
sns.pairplot(df, hue='sex')

# Display the plot
plt.title("Pairwise Plot: Tips Dataset")
plt.show()
```

```python
# Regression plot showing the relationship between total_bill and tip
sns.regplot(x='total_bill', y='tip', data=df)

# Display the plot
plt.title("Regression Plot: Total Bill vs Tip")
plt.show()
```

```python
# Distribution plot showing the distribution of total bill
sns.histplot(df['total_bill'], kde=True)

# Display the plot
plt.title("Distribution Plot: Total Bill")
plt.show()
```
